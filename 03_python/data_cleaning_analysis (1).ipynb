```python
import pandas as pd
#load data
df = pd.read_csv("Sample - Superstore.csv" , encoding='latin1')
#head
df.head()
```

```python
#basic info
df.info()
df.describe()
```

```python
#cleaning
df = df.drop_duplicates()                                        #remove dublicates
df = df.dropna()                                                 #remove missing value
df.columns = df.columns.str.strip().str.lower().str.replace(" ","_")     #clean coloumn name
print(df.columns)
df.head()
```

```python
#analysis
df.groupby('region')['sales'].sum()
```

```python
#visulization
import matplotlib.pyplot as plt

df.groupby('region')['sales'].sum().plot(kind='bar')
plt.title('sales by region')
plt.show()

```

```python
import matplotlib.pyplot as plt

df['order_date'] = pd.to_datetime(df['order_date'])

sales_trend = df.groupby('order_date')['sales'].sum()

sales_trend.plot()
plt.title('Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
# Sales vs profit
plt.scatter(df['sales'], df['profit'])

plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')

plt.show()
```
